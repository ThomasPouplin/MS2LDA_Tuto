For different pairs of adduct transformations at certain dilution level, are there any compounds that are detected together?

```python
def get_cooccurence(trans_results, t1, t2, l):
    t1mat = trans_results[t1]
    t2mat = trans_results[t2]
    t1mat = t1mat[:, l]
    t2mat = t2mat[:, l]
    t1mat = t1mat[:, None]
    t2mat = t2mat[:, None]
    results = np.dot(t1mat, t2mat.T)
    return results
```

```python
t1 = 6 # M+H
t2 = 10 # M+Na
print "Co-occurence of compounds in " + str(transformations[t1].name) + " vs " + \
    str(transformations[t2].name)
for l in range(L):
    results = get_cooccurence(trans_results, t1, t2, l)
    plt.spy(results)
    plt.ylabel("molDB (" + str(transformations[t1].name) + ")")
    plt.xlabel("molDB (" + str(transformations[t2].name) + ")")
    plt.title("Dilution " + dilutions[l])
    plt.show()
```
