We want to find out interesting stuff from the dilution data. First, we reproduce the heatmap on the spreadsheet. This plots for each transformation, the intensities of the detected adduct peaks across all dilution levels.

```python
dilutions = ["1:1", "1:5", "1:10", "1:50", "1:100", "1:1000"]

loader = FileLoader()
transformations = loader.load_transformation('mulsubs/mulsub2.txt')
database = loader.load_database('database/std1_mols.csv')

D = len(database)
L = len(dilutions)
T = len(transformations)
transformations_map = {}
for t in range(T):
    trans = transformations[t]
    transformations_map[trans] = t
```

```python
trans_results = {}
trans_masses = {}
for t in range(len(transformations)):
    
    print transformations[t]
    
    # filename of the previously stored results for each transformation t
    fname = './transformations_data/std1/' + str(t) + '.txt'

    # results is I by L, where I is the no. of molecules in database and 
    # L is the no. of dilution levels
    results = np.loadtxt(fname)

    fname = './transformations_data/std1/' + str(t) + '.mass.txt'    
    masses = np.loadtxt(fname)
    
    # make heatmap
    logged = np.log2(results) # log base 2 to follow the spreadsheet
    logged[logged==-np.inf]=0 # replace -inf with 0
    sres = logged[logged[:, 0].argsort()] # sort by first column
    heatmap = plt.pcolor(sres)
    plt.xlabel('dilutions')
    plt.ylabel('molDB')
    plt.title("Detected peaks log intensities (" + transformations[t].name + ")")
    plt.show()
    
    trans_results[t] = results
    trans_masses[t] = masses
```

We see from each heatmap that the majority of adduct transformations do not have any detected peaks. Here's the count for each detected adduct peaks across all dilution files:

```python
from collections import Counter
trans_freq = Counter()
for t in range(len(transformations)):
    results = trans_results[t]
    nnz = np.count_nonzero(results)
    trans_freq[transformations[t]] = nnz
for (key, val) in trans_freq.most_common():
    print key.name + " = " + str(val)
```

So, we can focus on the top few most common adducts, e.g. M+H, M+Na, M+2NA-H, M+K, etc. 

We can also plot detected compounds vs adducts for each dilution level.

```python
dilution_intenses = {} # maps from dilution l to the compound intensity across adducts matrix
dilution_masses = {} # maps from dilution l to the compound mass across adducts matrix
dilution_colours = ['b', 'g', 'r', 'k', 'y', 'm']

def populate_matrix():
    # populate dilution_intenses and dilution_masses matrices
    for l in range(L):
        results = np.zeros((D, T))
        masses = np.zeros((D, T))
        for d in range(D):
            for t in range(T):
                temp = trans_results[t]
                results[:, t] = temp[:, l]
                temp = trans_masses[t]
                masses[:, t] = temp[:, l]
        logged = np.log2(results) # log base 2 to follow the spreadsheet
        logged[logged==-np.inf]=0 # replace -inf with 0
        dilution_intenses[l] = logged
        dilution_masses[l] = masses    

def plot_adducts(n, show_mz=False):

    fig, axs = plt.subplots(1, L, sharex=False, sharey=True)
    
    for l in range(L):
        ax = axs[l]
        logged = dilution_intenses[l]
        logged = logged[0:n, :]
        ax.matshow(logged, aspect='equal')
        ax.set_title(dilutions[l]) 
        if l == 0:
            ax.set_ylabel("molDB")
        ax.set_xlabel("trans")
        ax.set_aspect('equal')
    plt.tight_layout()
    plt.show()

    if show_mz:
        fig, axs = plt.subplots(1, L, sharex=False, sharey=True)
        fig.set_size_inches(20, 2)
        for l in range(L):
            ax = axs[l]
            c = dilution_colours[l]
            logged = dilution_intenses[l]
            logged = logged[0:n, :]
            masses = dilution_masses[l]
            masses = masses[0:n, :]
            for (i,t), intensity in numpy.ndenumerate(logged):
                mass = masses[i, t]
                ax.plot((mass, mass), (0, intensity), c)
            ax.set_title(dilutions[l]) 
            if l == 0:
                ax.set_ylabel("log intensity")
            ax.set_xlabel("m/z")
        plt.tight_layout()
        plt.show()
```

```python
populate_matrix()
plot_adducts(D)
```

This shows that across the different dilutions, there are fewer detected compounds. And for those compounds initially detected in 1:1, we don't see as many adducts in say 1:1000. See below plot where we zoom to the first 40 compounds.

```python
plot_adducts(40, show_mz=True) # zoom into the first few compounds
```

For each compound, we plot the top-n detected adducts across dilutions. 

```python
def plot_compound(adducts):
    for i in range(D):
        db_entry = database[i]
        has_plot = False
        intenses_map = {}
        for trans in adducts:
            t = transformations_map[trans]
            all_ls = range(L)
            all_intenses = []
            for l in all_ls:
                c = dilution_colours[l]
                logged = dilution_intenses[l]
                intensity = logged[i, t]
                all_intenses.append(intensity)
            if np.sum(all_intenses)>0:
                plt.plot(all_ls, all_intenses, label=trans.name)
                has_plot = True
                intenses_map[trans] = all_intenses
        if has_plot:
            print db_entry
            for trans in adducts:
                if trans in intenses_map:
                    intensities = intenses_map[trans]
                    formatted_intensities = ['{:.3f}'.format(intensity) for intensity in intensities]
                    print "- " + trans.name + " = [" + ", ".join(formatted_intensities) + "]"                                           
            plt.title(db_entry.name + " (" + db_entry.formula + ")") 
            plt.ylabel("log intensity")
            plt.xlabel("dilutions")
            fontP = FontProperties()
            fontP.set_size('small')
            plt.legend(loc='upper right', prop = fontP)        
            plt.xticks(all_ls, dilutions)        
            plt.tight_layout()
            plt.show()

n = 5
print "Top " + str(n) + " most common adducts"
adducts = []
for (key, val) in trans_freq.most_common(n):
    print key.name + " = " + str(val)
    adducts.append(key)
print
plot_compound(adducts)
```

We can also maybe check which transformation pairs tend to occur together? 

```python
def get_cooccurence(trans_results, t1, t2, l):
    t1mat = trans_results[t1]>0
    t2mat = trans_results[t2]>0
    t1mat = t1mat[:, l]
    t2mat = t2mat[:, l]
    t1mat = t1mat[:, None]
    t2mat = t2mat[:, None]
    results = np.dot(t1mat, t2mat.T)
    return results

def plot_cooccurence(t1, t2):
    print "Co-occurence of compounds having " + str(transformations[t1].name) + " vs " + \
        str(transformations[t2].name + " adduct peaks in the same dilution file")
    for l in range(L):
        results = get_cooccurence(trans_results, t1, t2, l)
        plt.spy(results, markersize=1)
        plt.ylabel("molDB (" + str(transformations[t1].name) + ")")
        plt.xlabel("molDB (" + str(transformations[t2].name) + ")")
        plt.title("Dilution " + dilutions[l])
        plt.show()
```

So, in the plot below, the matrix entry at (i, j) is 1 if: the M+H adduct peak of the i-th compound is detected in the dilution data together with the M+Na adduct peak of the j-th compound.

```python
plot_cooccurence(6, 10)
```
