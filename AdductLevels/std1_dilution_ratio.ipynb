Just an example notebook to plot some ratio in the detected intensities between M+H and M+Na. First, load the preprocessed CSV data. This was previously done in dilution_export.ipynb, so see that notebook to change the preprocessing parameters etc.

```python
plotter = DilutionPlotter('std1')
plotter.populate_matrix()
```

We can extract some stuff from the plotter object

```python
D = plotter.D # no. of molecules
L = plotter.L # no. of dilutions
T = plotter.T # no. of transformations

for t in range(T):
    trans = plotter.transformations[t]
    print str(t) + ' ' + str(trans)
```

```python
xs = []
ys = []
for l in range(L):

    dil = plotter.dilutions[l]
    res = plotter.dilution_logged_results[l] # D x T matrix of the detected intensities
    for d in range(D):
        x = res[d, 6] # M+H
        y = res[d, 10] # M+Na
        xs.append(x)
        ys.append(y)
    c = plotter.dilution_colours[l]
    plt.plot(xs, ys, marker='.', linestyle='None', color=c)
    plt.title("Logged intensities of detected compounds -- dilution " + dil)
    plt.xlabel('M+H')
    plt.ylabel('M+Na')
    plt.show()
    
    # other useful things?
    # mm = plotter.dilution_masses[l] # D x T matrix of the detected masses
```
