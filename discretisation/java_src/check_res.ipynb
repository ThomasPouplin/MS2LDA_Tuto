```python
import pylab as plt
%matplotlib inline
import numpy as np
mass = []
rt = []
intensity = []
cluster = []
rel = []
trans = []
with open('testtxt.txt','r') as file:
    file.readline()
    for line in file:
        split_line = line.split()
        mass.append((float)(split_line[0]))
        rt.append((float)(split_line[1]))
        intensity.append((float)(split_line[2]))
        cluster.append((int)(split_line[3]))
        trans.append(split_line[4])
cluster = np.array(cluster)
mass = np.array(mass)
intensity = np.array(intensity)
rt = np.array(rt)
nPeaks = cluster.size
```

```python
import numpy as np
for clus_no in np.arange(nPeaks):
    pos = np.where(cluster == clus_no)[0]
    if pos.size > 4:
        print clus_no
        plt.figure()
        for p in pos:
            plt.plot([mass[p],mass[p]],[0,intensity[p]])
            plt.text(mass[p],intensity[p],trans[p])
```
