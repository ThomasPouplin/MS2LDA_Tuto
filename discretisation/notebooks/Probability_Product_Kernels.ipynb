<h2>Probability product kernels</h2>
<p>PPKs allow us to compute a similarity measure between two probability densities. For example, consider two Gaussians:
$$ y\sim {\cal N}(x_1,\sigma^2)~~~y\sim {\cal N}(x_1,\sigma^2) $$
We can compute the similarity by multiplying them together and integrating the result over all possible values of $y$:
$$ PPK = \int {\cal N}(y|x_1,\sigma^2){\cal N}(y|x_1,\sigma^2) ~dy$$
For Gaussians, this expression can be evaluated in closed form. See the following section (equation 371) from the matrix cookbook:
</p>

<p>The product in our integral is therefore equal to a constant multiplied by another Gaussian:
$$ \int {\cal N}(y|x_1,\sigma^2){\cal N}(y|x_1,\sigma^2) ~dy = c\int {\cal N}(y|m,s^2)~ dy $$
We can ignore the Gaussian and the integral -- the integral of a Gaussian pdf must equal 1. So, we only need the constant. According to the matrix cookbook, this will be:
$$ c = {\cal N}(x_1|x_2,2\sigma^2) $$
which is straightforward to calculate and is directly related to the distance between the two original means $x_1,x_2$. Here's an example:</p>

```python
x_1 = 3
x_2 = 4
sigma_sq = 1
```

```python
import pylab as plt
%matplotlib inline
import numpy as np
```

```python
def normpdf(x,m,ss):
    const = 1.0/(np.sqrt(2.0*np.pi*ss))
    return const*np.exp(-(1.0/(2.0*ss))*(x-m)**2)
```

<p>Plot the Gaussians</p>

```python
plt.figure()
xvals = np.arange(-1,7,0.01)
plt.plot(xvals,normpdf(xvals,x_1,sigma_sq))
plt.plot(xvals,normpdf(xvals,x_2,sigma_sq))
PPK = normpdf(x_1,x_2,2*sigma_sq)
title = "PPK = " + str(PPK)
plt.title(title)
```

<p>In the case of clusters of peaks, we represent the cluster as a mixture of Gaussians, one at each peak. Each mixture component has equal weight. For example, with peaks at $x_1,x_2,\ldots,x_N$ in cluster 1 and peaks at $z_1,z_2,\ldots,z_M$ in cluster 2, our PPK is:
$$ PPK = \int \left[\frac{1}{N}\sum_n {\cal N}(y|x_n,\sigma^2)\right]\left[\frac{1}{M}\sum_m {\cal N}(y|z_m,\sigma^2)\right]~dy $$
This can be switched to a double sum over an integral:
$$ PPK = \frac{1}{MN}\sum_m\sum_n \int {\cal N}(y|x_n,\sigma^2){\cal N}(y|z_m,\sigma^2) ~dy $$
Where the integrand is identical to our previous expression. Therefore:
$$ PPK = \frac{1}{MN}\sum_m\sum_n {\cal N}(x_n|z_m,2\sigma^2) $$
Here's an example:</p>

```python
sigma_sq = 0.1
x = np.array([1.0,2.0,3.0,4.0])
z = np.array([1.2,3.2,4.1])[:,None]
plt.figure()
for i in x:
    plt.plot(xvals,normpdf(xvals,i,sigma_sq),'b')
for i in z:
    plt.plot(xvals,normpdf(xvals,i,sigma_sq),'r')

PPK = 0
for i in x:
    for j in z:
        PPK += normpdf(i,j,2*sigma_sq)
# PPK/=(x.size*z.size)
title = "PPK = " + str(PPK)
plt.title(title)
```

<p>Note that the normalising term ($NM$) might be a problem for us as it means that there is no preference for more matches. Try adding and removing terms from $x$ or $z$ - it doesn't systematically increase or decrease the score as we would like. I'm sure we can fix this by either removing this constant, or changing the generative assumption.</p>
