<p>Load the peaks</p>

```python
def num(s):
    try:
        return int(s)
    except ValueError:
        return float(s)

file = open('std1-file1.identified.csv','r')
heads = file.readline()
mass = []
intensity = []
rt = []
print heads
for line in file:
    splitline = line.split(',')
    mass.append(num(splitline[1]))
    rt.append(num(splitline[2]))
    intensity.append(num(splitline[3]))

file.close()
    
mass = np.array(mass)[:,None]
rt = np.array(rt)[:,None]
intensity = np.array(intensity)[:,None]
```

```python
molecule_ID = []
molecule_name = []
molecule_formula = []
molecule_mass = []
for line in open('std1_mols.csv'):
    splitline = line.split(',')
    molecule_ID.append(splitline[0])
    molecule_name.append(splitline[1])
    molecule_formula.append(splitline[2])
    molecule_mass.append(num(splitline[3]))
    
```

<p>Load the transformations</p>

```python
adduct_name = []
adduct_mul = []
adduct_sub = []
adduct_del = []

for line in open('mulsub2.txt'):
    splitline = line.split(',')
    adduct_name.append(splitline[0])
    adduct_sub.append(num(splitline[1]))
    adduct_mul.append(num(splitline[2]))
    adduct_del.append(num(splitline[3]))
    
adduct_mul = np.array(adduct_mul)[:,None]
adduct_sub = np.array(adduct_sub)[:,None]
adduct_del = np.array(adduct_del)[:,None]
```

<p>Define the mass and RT tolerance parameters</p>

```python
import scipy.sparse as sp
import numpy as np
```

```python
massTol = 2.0 
rtTol = 5
```

```python
def massMatch(mass,otherMasses,tol):
    return np.abs((mass-otherMasses)/mass)<tol*1e-6

def rtMatch(rt,otherRT,tol):
    return np.abs(rt-otherRT)<tol

protonPos = np.flatnonzero(np.array(adduct_name)=='M+H')
print protonPos
clusterPriorMassMean = (mass - adduct_sub[protonPos])/adduct_mul[protonPos] 
nPeaks = mass.size
possible = sp.lil_matrix((nPeaks,nPeaks),dtype=np.int)
tMass = sp.lil_matrix((nPeaks,nPeaks),dtype=np.float)
for p in np.arange(nPeaks):
    tr = (mass[p] - adduct_sub)/adduct_mul + adduct_del
    rtm = rtMatch(rt[p],rt,rtTol)
    itm = (intensity[p]<=intensity)
    for t in np.arange(adduct_sub.size):
        a = massMatch(tr[t],clusterPriorMassMean,massTol)
        pos = np.flatnonzero(rtm*a*itm)
        possible[p,pos] = t+1
        tMass[p,pos] = tr[t]
        
    
```
